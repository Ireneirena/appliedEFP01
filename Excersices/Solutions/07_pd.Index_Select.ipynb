## Exercise 7

Δημιουργήστε ένα DataFrame `df` που περιέχει τις στήλες `country` και `variety` των πρώτων 100 εγγραφών.

Συμβουλή: μπορείτε να χρησιμοποιήσετε `loc` ή `iloc`. Όταν εργάζεστε για την απάντηση σε αυτήν την ερώτηση και σε πολλές από αυτές που ακολουθούν, κρατήστε το παρακάτω:

> Το «iloc» χρησιμοποιεί το σχήμα ευρετηρίου Python stdlib, όπου περιλαμβάνεται το πρώτο στοιχείο της περιοχής και το τελευταίο εξαιρείται. Tο `loc` εμανίζει και τα included.

> Αυτό προκαλεί ιδιαίτερη σύγχυση όταν το DataFrame είναι μια απλή αριθμητική λίστα, π.χ. `0,...,1000`. Σε αυτήν την περίπτωση το `df.iloc[0:1000]` θα επιστρέψει 1000 καταχωρήσεις, ενώ το `df.loc[0:1000]` θα επιστρέψει 1001 !!! Για να λάβετε 1000 στοιχεία χρησιμοποιώντας το "loc", θα χρειαστεί να πάτε ένα χαμηλότερα και να ζητήσετε "df.iloc[0:999]".


```python
cols = ['country', 'variety']
df = reviews[cols].loc[:99]
df
```

```python
#or
cols_idx = [0, 8]
df_alternative = reviews.iloc[:100, cols_idx]
df_alternative
```

## Exercise 9

Δημιουργήστε ένα DataFrame `top_south_europe` που περιέχει όλες τις κριτικές με τουλάχιστον 95 βαθμούς (από τους 100) για κρασιά από την Ιταλία, την Ισπανία ή την Ελλάδα .

```python
top_south_europe = reviews.loc[
    (reviews.country.isin(['Italy', 'Greece', 'Spain']))
    & (reviews.points >= 90)]

top_south_europe
```
